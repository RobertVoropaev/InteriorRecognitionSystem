```python
%matplotlib inline

#System
import os
import shutil
import random

#Base
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import cv2
from skimage.io import imread, imshow, imsave
from keras.utils import to_categorical
```

```python
mask = imread("data/ADE20K_formated/train/mask/mask/00000263.png")
```

```python
imshow(mask)
```

```python
for i in range(0, 32):
    imshow(to_categorical(mask, num_classes=32)[:, :, i])
    plt.title(i)
    plt.show()
```

```python
for i in [0, 2, 3, 5, 7, 26]:
    imsave("/home/rv/DiplomSegmentation/mask" + str(i) +".png", to_categorical(mask, num_classes=32)[:, :, i])
```
