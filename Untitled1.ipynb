```python
import numpy as np
import cv2
from skimage.io import imshow
cap = cv2.VideoCapture(0)
while(True):
    ret, frame = cap.read()
    imshow(frame)
    if cv2.waitKey() & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
